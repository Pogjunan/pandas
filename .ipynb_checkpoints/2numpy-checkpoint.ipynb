```python
r[:3,:3,0] = 255   # red 는 0번채널
g[:3,2:,1] = 255 #green 은 1번 채널
b[2:,:,2] = 255  #blue 는 2번 채널
```

```python
plt.imshow(r+g+b)
```

```python
import PIL 
```

```python
import numpy as np

netflix = np.array(PIL.Image.open('넷플.jpg'),dtype=np.int64)


# 사진 돌리기 필요시 'ijk->jik' 연산 수행은 사진 돌리기용
#arr = np.random.rand(2, 3, 4)
#result = np.einsum('ijk->jik', arr)
```

```python
netflix
```

```python
netflix.shape
```

```python
plt.imshow(netflix)
```

```python
plt.imshow(netflix[200:400, 800:1100,:]) #Netflix 로고 강요
```

```python
# 파란 넷플릭스 만들기

blue_netflix = netflix*0 
blue_netflix[:,:,2] = netflix[:,:,2]
plt.imshow(blue_netflix)
```

```python
plt.imshow(blue_netflix[200:400, 800:1100,:])
```

```python
#green
green_netflix = netflix*0 
green_netflix[:,:,1] = netflix[:,:,1]
plt.imshow(green_netflix)
```

```python
# 모든 값에 루트를 취해보면 어떻게 바뀔까요?

root_netf = np.sqrt(netflix).astype(int)
plt.imshow(np.concatenate([netflix,root_netf],axis=1))
```

```python
import numpy as np

netflix = np.array(PIL.Image.open('넷플.jpg'),dtype=np.int64)/255
```

```python
root_netf = np.sqrt(netflix)
plt.imshow(np.concatenate([netflix,root_netf],axis=1))

# 더 밝게 변했습니다. 0.3 -(루트)-> 0.5.. # 1에 가까울 수록 밝아진다.
```

```python
# 후광넣기

shiny_netf = np.sqrt(netflix)*(netflix>0.7)+netflix*(netflix<=0.7)
plt.imshow(np.concatenate([netflix,shiny_netf],axis=1))
```

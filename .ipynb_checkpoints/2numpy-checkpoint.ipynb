### **1 Import**

```python
import numpy as np
import matplotlib.pyplot as plt
```

기본 플랏 사용법

```python
x=[1,2,3,4]
y=[1,2,4,3] 

plt.plot(x,y)
```

```python
# plt 에서 모양변경하는 가벼운 방법
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'r--')
```

```python
# 하나의 문법이라고 생각하기.
plt.plot(x,y,'b:') # 색상 + 선종류 야무지게 바뀔수있다.
```

```python
plt.plot(x,y,'ko') # k = 검정색(b=blue 라서 blac'k' 의 k 이용) , o = 큰 점
```

<br>회귀분석 및 머신러닝에서 많이 보일만한 그림


```python
x = np.arange(-5,5,0.1) #100가지
e = np.random.randn(100) 
y = 2*x + e
len(x)
```

```python
plt.plot(x,y,'.')
plt.plot(x,2*x,'--')
```

# 이미지 자료 다루기

```python
r = np.array([0]*25*3).reshape(5,5,3) 
g = np.array([0]*25*3).reshape(5,5,3) 
b = np.array([0]*25*3).reshape(5,5,3)
```

```python
r
```

```python
r[:,:,0], r[:,:,1], r[:,:,2]
```

```python
r[:3,:3,0] = 255   
```

```python
r[:,:,0], r[:,:,1], r[:,:,2]
```

```python
g[:,:,0], g[:,:,1], g[:,:,2]
```

```python
r[:3,:3,0] = 255   # red 는 0번채널
g[:3,2:,1] = 255 #green 은 1번 채널
b[2:,:,2] = 255  #blue 는 2번 채널
```

```python
plt.imshow(r+g+b)
```

```python
r[:3,:3,0] = 130
g[:3,2:,1] = 130
b[2:,:,2] = 130
```

```python
plt.imshow(r+g+b) #채도 낮추기
```

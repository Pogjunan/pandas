# 클래스란?

` 클래스는 함수를 정의하는 기존방식에서 발전된 방식으로 함수를 정의하고 여러 함수를 이을 수 있는 수단이다.`
- 클래스는 여러 함수를 잇고 용도에 맞게 자유자재로 사용 ,변경 및 활용을 할 수 있다.

`-` 함수 만들기 전 가볍게 그림 커스터마이징하기

````python
from IPython.core.display import HTML
````

````python
# html 형식으로 style 을 지정하고 만들고싶은대로 만든 그림입니다.

html_짱구 = '''
<style>
    .title {
        font-family: "Times New Roman", serif;
        font-size: 30px;
        font-weight: 900;
    }
    .text {
        font-family: "Arial", sans-serif;
        font-size: 20px;
        font-style: italic;
    }
    .highlight {
        font-family: "Montserrat", monospace;
        font-size: 35px;
        font-weight: 900;
        text-decoration: underline;
        font-style: normal;
        color: darkblue;
        background-color: #FFFF00;
    }
</style>
<p class="title">나는 짱구가 좋아.</p>
<img src=짱구그림1.jpg width="300">


'''
HTML(html_짱구)
````

## 함수 만들기
-----


````python
# 1 사용할 수 있게 틀 만들기.
````

````python
html_짱구 = '''
<style>
    .title {{
        font-family: "Times New Roman", serif;
        font-size: 30px;
        font-weight: 900;
    }}
    .text {{
        font-family: "Arial", sans-serif;
        font-size: 20px;
        font-style: italic;
    }}
    .highlight {{
        font-family: "Montserrat", monospace;
        font-size: 35px;
        font-weight: 900;
        text-decoration: underline;
        font-style: normal;
        color: darkblue;
        background-color: #FFFF00;
    }}
</style>

<p class="title">{ttl}</p>
<img src={url} width="300">
<p class="text"> {T}</p>
<p class="title">지금 당장 <span class="highlight">{h1}</span></p>
'''
````

````python
title = '나는 짱구가 좋아.'
imgurl = 'http://210.117.173.186:26239/files/study/%EA%B8%B0%EC%B4%88/DeeplearningwithPytorch/%EC%A7%B1%EA%B5%AC%EA%B7%B8%EB%A6%BC1.jpg\
?_xsrf=2%7Cea72de89%7Cba426c89f852e33ef85a83537fa133d7%7C1704794540'

txt = '훈이는 싫어.'
highlight = '초코비먹고싶다.'
````

````python
def show_짱구(html_짱구,title,txt,highlight):
    잠옷짱구 = html_짱구.format(
        ttl = title,
        url = imgurl,
        T = txt,
        h1 = highlight
    )
    display(HTML(잠옷짱구))
        
show_짱구(html_짱구,title,txt,highlight)
````

# 해석
<br>
`-` 짱구라는 함수를 만들었고 이 함수에 적당한 예시코드를 넣으면 자돋으로 실행되게 만드는 것이 중요하다.
<br>

`html_짱구` : 짱구 그림의 틀을 완성시키고 필요한 파라미터 변경과 같은 것들을 변수로 받게 제작(변수: title,txt,highlight)

<br>

`-` 잠옷짱구 : 잠옷짱구라고 이름을 지은 이유는 많은 예시 중 이번에 잠옷짱구의 예시를 쓸 것이라는 은유적 표현이다. 이는 다른 사람들이 자신의 코드를 볼 때 직관적으로 쉽게 만들 수 있는 협업에 중요한 능력이라고 볼 수 있다.

<br>

`-` 잠옷짱구 : 잠옷짱구는 html_짱구라는 html 식의 format 을 변수(인수)로 받아서 실행시킨다.

<br>

`-` show_짱구(html_짱구,title,txt,highlight) : 함수를 실행시키는 셀로 `show_짱구` 라는 미리 `def` 로 지정된 함수를 사용하여 괄호안에 있는 변수를 사용하여 제작하라는 의미입니다.

<br>

`-` 변수에 대해 : 이 변수들은 현재 ipynb 파일 내에서 `전역`에 포함되어 있는 변수입니다. 새로 함수를 제작하게되면 이 변수들은 사용할 수 없습니다. 그래서 사용할 수 있도록 하는 것이 괄호내에 변수명을 적는 것입니다.

### 해석
`-` `짱구` 클래스를 만들었고 `맹구` 는 이제 짱구 class 소속이라는 의미입니다.

`-` 맹구는 현재 변수입니다. 짱구라는 class 에 소속된 변수.

```
맹구.

보통 
```

해도 아무것도 나타나지 않습니다. <br>
클래스로 지정된 변수는 밑의 사진처럼 사용할 수 있는 함수를 나타나게 합니다.

````python
맹구.title = "맹구가 사실 더 좋아."
맹구.imgurl = '맹구돌컬렉션.jpg'
맹구.highlight = '초코비보다 돌을 모으고 싶 어.'
txt = '돌을 모으자..'
맹구.html_짱구 = '''
<style>
    .title {{
        font-family: "Times New Roman", serif;
        font-size: 30px;
        font-weight: 900;
    }}
    .text {{
        font-family: "Arial", sans-serif;
        font-size: 20px;
        font-style: italic;
    }}
    .highlight {{
        font-family: "Montserrat", monospace;
        font-size: 35px;
        font-weight: 900;
        text-decoration: underline;
        font-style: normal;
        color: darkblue;
        background-color: #FFFF00;
    }}
</style>

<p class="title">{ttl}</p>
<img src={url} width="300">
<p class="text"> {T}</p>
<p class="title">지금 당장 <span class="highlight">{h1}</span></p>
'''
````

## `맹구.` 을 치고
<br>tab 을 누르면 이렇게 떠서 함수가 사용이 가능해진다. 


````python
def show(맹구):
    _str = 맹구.html_짱구.format(
        ttl = 맹구.title,
        url = 맹구.imgurl,
        h1 = 맹구.highlight,
        T = txt
    )
    display(HTML(_str))
````

````python
show(맹구)
````

````python
# 맹구 대신 딴걸로하자. 짱아는 지금 짱구 class 에 저장된다.
짱아 = 짱구()
짱아
````

````python
짱아.title = "짱아를 버리면 안되지 "
짱아.imgurl = '짱아사진.jpg'
짱아.highlight = '땨땨땨땨'
짱아.txt = '땨땨땨땨땨!'
짱아.html_짱구 = '''
<style>
    .title {{
        font-family: "Times New Roman", serif;
        font-size: 30px;
        font-weight: 900;
    }}
    .text {{
        font-family: "Arial", sans-serif;
        font-size: 20px;
        font-style: italic;
    }}
    .highlight {{
        font-family: "Montserrat", monospace;
        font-size: 35px;
        font-weight: 900;
        text-decoration: underline;
        font-style: normal;
        color: darkblue;
        background-color: #FFFF00;
    }}
</style>

<p class="title">{ttl}</p>
<img src={url} width="300">
<p class="text"> {T}</p>
<p class="title">땨땨 <span class="highlight">{h1}</span></p>
'''
````

`짱아.` 을하면 똑같이 함수를 사용할 수 있다. <br>
`show()` 함수는 재사용 가능하다. <br>

````python
def show(맹구):
    _str = 맹구.html_짱구.format(
        ttl = 맹구.title,
        url = 맹구.imgurl,
        h1 = 맹구.highlight,
        T = 맹구.txt
    )
    display(HTML(_str))
````

```
show 안의 `맹구`라는 class 로 지정된 함수를 넣었는데
맹구 대신 짱아를 넣으면 모든 함수가 짱아에 대한 함수로 바껴서
적용이 잘 된다.
```

````python
짱아.highlight = '땨땨땨땨'

show(짱아)
````

````python
# 자유자재 변경

짱아.txt ='짱아야 짱구물어!'

show(짱아)
````

````python
class 짱구:
    pass 
````

`-` 아래의 과정은 `짱구`라는 이름의 클래스를 선언 <br>
`-` 하나의 짱구라는 폴더를 만들었다고 생각하면된다.

````python
맹구 = 짱구()
맹구
````

`-` 짱구의 등장인물 중 맹구가 짱구에 속해있다는 것을 알려주는 것으로
<br>
맹구는 짱구의 폴더 안에서 행동할 수 있다고 생각
<br>
`현재 맹구는 변수이다.`

````python
맹구.title = '짱구의 등장인물 맹구'
맹구.imgurl = '맹구돌컬렉션.jpg' # 파일의 위치로 넣기
맹구.txt = '맹구의 콧물어택'
맹구.highlight = '돌이 최고 마인드'
````

````python
맹구.html_짱구 =\
'''
<style>
    .title {{
        font-family: "Times New Roman", serif;
        font-size: 30px;
        font-weight: 900;
    }}
    .text {{
        font-family: "Arial", sans-serif;
        font-size: 20px;
        font-style: italic;
    }}
    .highlight {{
        font-family: "Montserrat", monospace;
        font-size: 35px;
        font-weight: 900;
        text-decoration: underline;
        font-style: normal;
        color: darkblue;
        background-color: #FFFF00;
    }}
</style>

<p class="title">{ttl}</p>
<img src={url} width="600">
<p> \n </p>
<p class="text"> {T}</p>
<p class="title">중요한 것은 꺾이지 않는 <span class="highlight">{h1}</span></p>
'''
````

````python
def show(맹구):
    _str = 맹구.html_짱구.format(
        ttl = 맹구.title,
        url = 맹구.imgurl,
        h1 = 맹구.highlight,
        T = 맹구.txt
    )
    display(HTML(_str))
````

````python
show(맹구)
````

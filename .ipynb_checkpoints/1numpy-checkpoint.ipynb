`-` np.concatenate 는 축의 총 갯수를 유지하면서 결합, np.stack은 축의 갯수를 하나 증가시키면서 결합의 차이가 있지만 현재는 패쓰

````python
# argmax , argmin
````

````python
a = np.random.rand(15).reshape(3,5)
a
````

````python
a = np.stack([np.array(l)/l.sum() for l in a]) 
# 리스트는 5개의 원소가 있으며 각 원소는 차원이 (5,) => stack을 이용해 첫 위치에 축을 추가하여 (3,5) 매트릭스를 만듬 
a # 각각 줄에서 자신의 크기에 대한 비율을 나타냄.
````

````python
np.argmax(a,axis=1) # 0 번째,2번째 , 3번째가 가장 큰 값을 가짐.
````

````python
np.argmin(a,axis=1)
````
